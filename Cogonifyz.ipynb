# Task 6 Savings Objectives

```python
# Step 1: Analyze Savings Objectives
objective_counts = df['Objective'].value_counts()
print("Main Savings Objectives (Frequency):\n")
print(objective_counts)
```

```python
# Step 2: List and Describe Objectives
print("\nDescriptions of Main Savings Objectives:\n")
descriptions = {
    "Capital Appreciation": "Growing the value of invested money over time.",
    "Retirement": "Saving funds for financial security after retirement.",
    "Growth": "Achieving long-term wealth growth.",
    "Education": "Saving for education expenses (self or children).",
    "Health Care": "Setting aside money for medical or health-related expenses.",
    "Wealth Creation": "Building overall wealth through investments.",
    "Liquidity": "Maintaining easy access to cash when required.",
    "Better Future": "Ensuring financial stability and improvement in lifestyle."
}
for obj, count in objective_counts.items():
    desc = descriptions.get(obj, "General savings/financial goal.")
    print(f"- {obj}: {desc} (Chosen by {count} participants)")
```

```python
# Optional: Visualization
plt.figure(figsize=(8,5))
objective_counts.plot(kind='bar', color='darkblue')
plt.title("Savings Objectives of Participants")
plt.xlabel("Objective")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

# Task 8  Investment Duration

```python
# Check unique values in Duration column
print("Unique Duration values:", df['Duration'].unique())
```

```python
# Step 2: Map duration ranges to numeric midpoints (example mapping, adjust as per dataset)
Duration_mapping = {
    'Less than 1 year': 0.5,
    '1-3 years': 2,
    '3-5 years': 4,
    '5-10 years': 7.5,
    'Above 10 years': 12  # You can adjust this assumption
}
```

```python
# Apply mapping
df['Duration_numeric'] = df['Duration'].map(Duration_mapping)
```

```python
# Step 3: Calculate average investment duration
average_duration = df['Duration_numeric'].mean()

print("\nAverage Investment Duration (in years):", round(average_duration, 2))
```
